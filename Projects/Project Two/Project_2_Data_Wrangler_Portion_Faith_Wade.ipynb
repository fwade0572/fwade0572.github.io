# Combining the Datasets

Combining both the Red & White Wine Datasets, as well as putting each value into their designated columns versus its original state.

```python
# Importing the necessary applications:
from google.colab import files
uploaded = files.upload()
import pandas as pd
# Reading the datasets:
red = pd.read_csv("winequality-red.csv", sep = ";")
white = pd.read_csv("winequality-white.csv", sep = ";")
```

```python
# Adding the wine type columns:
red["wine_type"] = "red"
white["wine_type"] = "white"
```

```python
# Combining & cleaning the datasets with their columns for further accuracy.
wine_df = pd.concat([red, white], ignore_index = True)
wine_df.columns = wine_df.columns.str.strip().str.replace(" ", "_")
```

```python
# Reviewing new combined dataset:
print("Shape:", wine_df.shape)
print(wine_df.head())
```

```python
# Saving the dataset & downloading it for further wrangling:
wine_df.to_csv("combined_wine_quality.csv")
files.download("combined_wine_quality.csv")
print("Dataset has been combined & downloaded successfully!")
```

# Cleaning the New Combined Dataset

Removing missing values as well as duplicates in order to prevent any unknown values from manipulating the data for later analyzation from the Data Scientist & Visualizer.

```python
# Removing duplicates and showing values before and after as a way to check for any remaining:
print("Duplicates Before:", wine_df.duplicated().sum())
wine_df = wine_df.drop_duplicates()
print("Duplicates After:", wine_df.duplicated().sum())
```

```python
# Although the duplicates have already likely taken care of the missing values, taking further measures to ensure there are no errors:
print("\nMissing Values Before:\n", wine_df.isnull().sum())
wine_df = wine_df.dropna()
print("\nMissing Values After:\n", wine_df.isnull().sum())
```

```python
# Additionally fixing the data types for the values designated categories:
wine_df["wine_type"] = wine_df["wine_type"].astype("category")
```

# Transforming the Wine Quality Dataset
Encoding categorical variables, scaling, and feature engineering the dataset in order to have a more effective and clearer visual for more efficient results.

```python
# Encoding the categorical variables through binary means: red is 0, white is 1.
wine_df["wine_type_encoded"] = wine_df["wine_type"].map({"red": 0, "white": 1})
```

```python
# Scaling the dataset to ensure future spitting, training, and testing of the data flows smoothly:
import numpy as np
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numeric_cols = wine_df.select_dtypes(include = np.number).columns
wine_df[numeric_cols] = scaler.fit_transform(wine_df[numeric_cols])
```

```python
# Adding a feature engineering attribution for simpler analyzation of alcohol levels:
wine_df["alcohol+level"] = pd.cut(wine_df["alcohol"], bins = 3, labels = ["low", "medium", "high"])
wine_df.head()
```

# Assessing & Finalizing the Cleaned Dataset
Importing summary statistics, value counts, and checking the final shape to ensure there are no further errors and/or miscalculations.

```python
# Importing Summary Statistics:
wine_df.describe()
```

```python
# Importing Value Counts:
print("\nWine Type Distribution:\n", wine_df["wine_type"].value_counts())
```

```python
# Importing the finalized shape of the data:
print("\nFinal Shape:", wine_df.shape)
wine_df.head()
```

```python
# Saving & Downloading the Finalized Cleaned Dataset:
wine_df.to_csv("clean_wine_quality.csv", index = False)
files.download("clean_wine_quality.csv")
```
